## Import Data and libraries

```python
import pandas as pd




# Load the TSV data
file_path = 'correct_twitter_201904.tsv'
twitter_data = pd.read_csv(file_path, sep='\t')
twitter_data.head()
```

### this Scrpt to search for a specific term like "britney_201904" in the dataset


```python
def search_term_in_tweets(df, term):
    term = term.lower()
    search_results = df[
        df['hashtags'].str.contains(term, case=False, na=False) |
        df['urls'].str.contains(term, case=False, na=False) |
        df['event'].str.contains(term, case=False, na=False)
    ]
    
    return search_results

results = search_term_in_tweets(twitter_data, 'britney_201904')

# Display the number of matching results and the first few rows
info = results.shape[0], results.head()
print(info)
```

### I have updated my script to include an input function that makes it simple to search through the data set for anything. ### I worked on six things that you can see in this code.

```python
import pandas as pd

def search_term_analysis(df, term):
# Ensure the search term is in lowercase
    term = term.lower()

# Filter tweets that contain the search term in relevant fields
    filtered_tweets = df[
        df['hashtags'].str.contains(term, case=False, na=False) |
        df['urls'].str.contains(term, case=False, na=False) |
        df['event'].str.contains(term, case=False, na=False)
    ]
    
# 1. Number of tweets posted containing the term on each day
    filtered_tweets['ts1'] = pd.to_datetime(filtered_tweets['ts1'])  # Convert to datetime if not already
    tweets_per_day = filtered_tweets.groupby(filtered_tweets['ts1'].dt.date).size()

# 2. Number of unique users who posted tweets containing the term
    unique_users = filtered_tweets['author_id'].nunique()

# 3. Average number of likes for tweets containing the term
    if 'like_count' in df.columns:
        avg_likes = filtered_tweets['like_count'].mean()
    else:
        avg_likes = "N/A - like count not available"

# 4. Place IDs where tweets came from
    place_ids = filtered_tweets['place_id'].unique() if 'place_id' in df.columns else "N/A - place ID not available"

# 5. Times of day tweets were posted
    times_of_day = filtered_tweets['ts1'].dt.hour.value_counts().sort_index()

# 6. User who posted the most tweets containing the term
    most_active_user = filtered_tweets['author_id'].value_counts().idxmax()

 # Compile results into a dictionary for easy output
    result = {
        "tweets_per_day": tweets_per_day,
        "unique_users": unique_users,
        "avg_likes": avg_likes,
        "place_ids": place_ids,
        "times_of_day": times_of_day,
        "most_active_user": most_active_user
    }
    
    return result

# Get search term from user
search_term = input("Enter the term you want to analyze: ")

# Perform the search and analysis
analysis_results = search_term_analysis(twitter_data, search_term)


# Display the results
print(f"Number of tweets per day:\n{analysis_results['tweets_per_day']}")
print(f"Number of unique users: {analysis_results['unique_users']}")
print(f"Average likes: {analysis_results['avg_likes']}")
print(f"Place IDs: {analysis_results['place_ids']}")
print(f"Times of day tweets were posted:\n{analysis_results['times_of_day']}")
print(f"Most active user: {analysis_results['most_active_user']}")
```
